# LTEE Mutation Temporal Analysis

## Project Goal
Test the module-selection balance prediction from Kim et al.: beneficial mutations should be concentrated in few genes/operons early in evolution and become more dispersed later as selection coefficients equilibrate across functional modules.

## Data Source
- **Paper**: Good BH, McDonald MJ, Barrick JE, Lenski RE, Desai MM. "The Dynamics of Molecular Evolution Over 60,000 Generations." *Nature* 551:45-50 (2017)
- **Repository**: https://github.com/benjaminhgood/LTEE-metagenomic
- **Reference genome**: REL606 (NC_012967.1)

## Key Analysis Choices

**Why exclude intergenic mutations?**  
We are testing whether mutations concentrate in functional modules (genes/operons). Intergenic mutations cannot be assigned to a specific gene or operon, so they are excluded from this analysis. This is consistent with Good et al.'s gene-level and operon-level parallelism analyses.

**Why t* = 17,500 generations?**  
Good et al. identified this as the partition time that best separates "coupon-collecting" genes (hit early, spread across populations) from "historically contingent" genes (hit later, clustered in specific populations). We use this as our early/late cutoff.

**Why Simpson's D?**  
Simpson's D measures the probability that two randomly chosen mutations hit the same gene/operon. It is sensitive to highly-hit targets (due to squaring proportions), making it ideal for detecting concentration in a few "hot" genes.

# Part 1: Setup and Data Loading

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
from Bio import SeqIO
from scipy.stats import entropy as scipy_entropy

# Paths
LTEE_REPO_PATH = Path('../LTEE-metagenomic')
LTEE_DATA_PATH = LTEE_REPO_PATH / 'data_files'
LTEE_ADDITIONAL_PATH = LTEE_REPO_PATH / 'additional_data'

# The six nonmutator populations
NONMUTATOR_POPULATIONS = ['m5', 'm6', 'p1', 'p2', 'p4', 'p5']

# Key time threshold from Good et al.
T_STAR = 17500

print("Setup complete")
print(f"  Data path: {LTEE_DATA_PATH}")
print(f"  Populations: {NONMUTATOR_POPULATIONS}")
print(f"  Early/late cutoff: {T_STAR} generations")
```

## Load HMM States and Build Mutation Table

Each mutation has an "appearance time" — the first timepoint where the HMM detected it as present (state > 0). We extract this from the `*_well_mixed_state_timecourse.txt` files.

```
def load_hmm_states(population):
    """Load HMM state trajectories for a population."""
    filepath = LTEE_DATA_PATH / f'{population}_well_mixed_state_timecourse.txt'
    with open(filepath, 'r') as f:
        lines = f.readlines()
    timepoints = [int(float(x)) for x in lines[0].strip().split(', ')]
    states = []
    for line in lines[5:]:
        vals = [int(float(x)) for x in line.strip().rstrip(',').split(', ') if x]
        states.append(vals)
    return timepoints, states

def get_appearance_time(state_trajectory, timepoints):
    """Find first timepoint where mutation appears (state > 0)."""
    for i, state in enumerate(state_trajectory):
        if state > 0:
            return timepoints[i]
    return None

def build_mutation_table(populations):
    """Build master table with all mutations and appearance times."""
    all_data = []
    for pop in populations:
        # Load annotation file
        filepath = LTEE_DATA_PATH / f'{pop}_annotated_timecourse.txt'
        full_df = pd.read_csv(filepath, header=0)
        full_df.columns = full_df.columns.str.strip()
        full_df['Passed?'] = full_df['Passed?'].str.strip()
        
        # Filter to PASS only and reset index
        pass_df = full_df[full_df['Passed?'] == 'PASS'].reset_index(drop=True)
        
        # Load HMM states
        timepoints, states = load_hmm_states(pop)
        
        # Build table
        for i, row in pass_df.iterrows():
            if i >= len(states):
                continue
            appearance_time = get_appearance_time(states[i], timepoints)
            all_data.append({
                'population': pop,
                'position': row['Position'],
                'gene': row['Gene'].strip() if isinstance(row['Gene'], str) else row['Gene'],
                'allele': row['Allele'].strip() if isinstance(row['Allele'], str) else row['Allele'],
                'annotation': row['Annotation'].strip() if isinstance(row['Annotation'], str) else row['Annotation'],
                'appearance_time': appearance_time
            })
    return pd.DataFrame(all_data)

# Build the master table
master_df = build_mutation_table(NONMUTATOR_POPULATIONS)
print(f"Total mutations loaded: {len(master_df)}")
```

## Preprocessing: Exclude Intergenic and Map to Operons

**Intergenic mutations** are excluded because they cannot be assigned to a gene or operon. This is consistent with Good et al.'s approach for gene-level analysis.

**Operon mapping** uses the DOOR database (door_operon_list.txt) to group genes into operons — co-transcribed gene units that often function together.

```
# Step 1: Exclude intergenic mutations
df = master_df[master_df['gene'] != 'intergenic'].copy()
print(f"After excluding intergenic: {len(df)} mutations ({len(master_df) - len(df)} removed)")

# Step 2: Build gene-to-operon mapping
operon_file = LTEE_ADDITIONAL_PATH / 'door_operon_list.txt'
operon_raw = pd.read_csv(operon_file, sep='\t')
locus_to_operon = dict(zip(operon_raw['Synonym'], operon_raw['OperonID']))

gbk_file = LTEE_ADDITIONAL_PATH / 'REL606.6.gbk'
gene_to_locus = {}
for record in SeqIO.parse(gbk_file, 'genbank'):
    for feature in record.features:
        if feature.type in ['CDS', 'gene']:
            locus_tag = feature.qualifiers.get('locus_tag', [None])[0]
            gene_name = feature.qualifiers.get('gene', [None])[0]
            if locus_tag and gene_name:
                gene_to_locus[gene_name] = locus_tag

def get_operon(gene_name):
    """Map gene name to operon ID."""
    if gene_name in locus_to_operon:
        return locus_to_operon[gene_name]
    locus = gene_to_locus.get(gene_name)
    if locus and locus in locus_to_operon:
        return locus_to_operon[locus]
    return None

# Add operon column
df['operon'] = df['gene'].apply(get_operon)
n_with_operon = df['operon'].notna().sum()
print(f"Mutations with operon mapping: {n_with_operon} / {len(df)} ({100*n_with_operon/len(df):.1f}%)")
```

## Data Summary

```
print("="*60)
print("FINAL DATASET SUMMARY")
print("="*60)
print(f"\nTotal genic mutations: {len(df)}")
print(f"Unique genes: {df['gene'].nunique()}")
print(f"Unique operons: {df['operon'].nunique()}")
print(f"\nAppearance time range: {df['appearance_time'].min()} - {df['appearance_time'].max()} generations")
print(f"\nMutations per population:")
print(df['population'].value_counts().to_string())
print(f"\nAnnotation types:")
print(df['annotation'].value_counts().to_string())
```

# Part 2: Early vs Late Analysis

We split mutations into **early** (≤17,500 gen) and **late** (>17,500 gen) periods based on Good et al.'s partition time.

```
# Split data
early = df[df['appearance_time'] <= T_STAR].copy()
late = df[df['appearance_time'] > T_STAR].copy()

# For operon analysis, filter to mutations with operon mapping
early_op = early[early['operon'].notna()].copy()
late_op = late[late['operon'].notna()].copy()

print(f"Early (≤{T_STAR} gen): {len(early)} mutations, {early['gene'].nunique()} genes")
print(f"Late (>{T_STAR} gen): {len(late)} mutations, {late['gene'].nunique()} genes")
```

## Top Genes by Period

Which genes are most frequently mutated in each period?

```
print("Top 10 genes - EARLY:")
print(early['gene'].value_counts().head(10).to_string())
print(f"\nTop 10 genes - LATE:")
print(late['gene'].value_counts().head(10).to_string())
```

# Part 3: Simpson's D Analysis

**Simpson's D** measures the probability that two randomly chosen mutations hit the same gene (or operon). 

- **Higher D** = mutations concentrated in fewer targets
- **Lower D** = mutations spread across many targets

**Hypothesis**: Early mutations should have higher D (more concentrated), late mutations should have lower D (more dispersed).

```
def simpson_d(labels):
    """
    Calculate Simpson's D: probability that two random picks are the same category.
    D = sum(p_i^2) where p_i is the proportion in category i.
    """
    counts = pd.Series(labels).value_counts()
    proportions = counts / counts.sum()
    return (proportions ** 2).sum()

# Calculate Simpson's D
early_gene_D = simpson_d(early['gene'])
late_gene_D = simpson_d(late['gene'])
early_operon_D = simpson_d(early_op['operon'])
late_operon_D = simpson_d(late_op['operon'])

print("Simpson's D (probability two random mutations hit same target):")
print(f"  Gene-level:   Early = {early_gene_D:.4f}, Late = {late_gene_D:.4f}")
print(f"  Operon-level: Early = {early_operon_D:.4f}, Late = {late_operon_D:.4f}")
```

## Permutation Test

The late period has more mutations than early. Could the lower Simpson's D in late simply be due to larger sample size?

**Permutation test**: Randomly shuffle all mutations into "early" and "late" groups (keeping original sizes), recalculate the difference in D. Repeat 10,000 times. If the observed difference is larger than all permuted differences, then the effect is real (not just sample size).

```
# Permutation test
n_perm = 10000
n_early = len(early)
n_early_op = len(early_op)

all_genes = df['gene'].values
all_operons = df[df['operon'].notna()]['operon'].values

obs_diff_gene = early_gene_D - late_gene_D
obs_diff_operon = early_operon_D - late_operon_D

perm_diffs_gene = []
perm_diffs_operon = []

np.random.seed(42)  # For reproducibility
for _ in range(n_perm):
    # Gene level
    shuffled = np.random.permutation(all_genes)
    d_early = simpson_d(shuffled[:n_early])
    d_late = simpson_d(shuffled[n_early:])
    perm_diffs_gene.append(d_early - d_late)
    
    # Operon level
    shuffled_op = np.random.permutation(all_operons)
    d_early_op = simpson_d(shuffled_op[:n_early_op])
    d_late_op = simpson_d(shuffled_op[n_early_op:])
    perm_diffs_operon.append(d_early_op - d_late_op)

perm_diffs_gene = np.array(perm_diffs_gene)
perm_diffs_operon = np.array(perm_diffs_operon)

# P-values (one-tailed: observed diff should be larger than permuted)
p_gene = np.mean(perm_diffs_gene >= obs_diff_gene)
p_operon = np.mean(perm_diffs_operon >= obs_diff_operon)

print(f"Permutation test (n={n_perm}):")
print(f"  Gene-level:   observed diff = {obs_diff_gene:.4f}, p = {p_gene:.4f}")
print(f"  Operon-level: observed diff = {obs_diff_operon:.4f}, p = {p_operon:.4f}")
```

```
# Plot Simpson's D results
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# Top left: Gene Simpson's D
ax1 = axes[0, 0]
bars1 = ax1.bar([f'Early (≤{T_STAR//1000}k)\nn={len(early)}', 
                 f'Late (>{T_STAR//1000}k)\nn={len(late)}'], 
                [early_gene_D, late_gene_D], color=['steelblue', 'darkorange'])
ax1.set_ylabel("Simpson's D")
ax1.set_title('Gene-Level')
for bar, val in zip(bars1, [early_gene_D, late_gene_D]):
    ax1.text(bar.get_x() + bar.get_width()/2, val + 0.0002, f'{val:.4f}', ha='center', fontsize=11)

# Top right: Operon Simpson's D
ax2 = axes[0, 1]
bars2 = ax2.bar([f'Early (≤{T_STAR//1000}k)\nn={len(early_op)}', 
                 f'Late (>{T_STAR//1000}k)\nn={len(late_op)}'], 
                [early_operon_D, late_operon_D], color=['steelblue', 'darkorange'])
ax2.set_ylabel("Simpson's D")
ax2.set_title('Operon-Level')
for bar, val in zip(bars2, [early_operon_D, late_operon_D]):
    ax2.text(bar.get_x() + bar.get_width()/2, val + 0.0002, f'{val:.4f}', ha='center', fontsize=11)

# Bottom left: Gene permutation null
ax3 = axes[1, 0]
ax3.hist(perm_diffs_gene, bins=50, color='gray', alpha=0.7, edgecolor='black')
ax3.axvline(x=obs_diff_gene, color='red', linewidth=2, label=f'Observed = {obs_diff_gene:.4f}')
ax3.set_xlabel("Difference in Simpson's D (Early - Late)")
ax3.set_ylabel('Count')
ax3.set_title(f'Permutation Test (p = {p_gene:.4f})')
ax3.legend()

# Bottom right: Operon permutation null
ax4 = axes[1, 1]
ax4.hist(perm_diffs_operon, bins=50, color='gray', alpha=0.7, edgecolor='black')
ax4.axvline(x=obs_diff_operon, color='red', linewidth=2, label=f'Observed = {obs_diff_operon:.4f}')
ax4.set_xlabel("Difference in Simpson's D (Early - Late)")
ax4.set_ylabel('Count')
ax4.set_title(f'Permutation Test (p = {p_operon:.4f})')
ax4.legend()

plt.suptitle("Simpson's D: Early vs Late Mutations\n(Higher D = more concentrated in fewer targets)", fontsize=14)
plt.tight_layout()
plt.show()
```

## Interpretation

**Results**: Early mutations have significantly higher Simpson's D than late mutations at both gene and operon levels (p < 0.0001). This means:

- **Early**: If you pick two random mutations, there's a higher probability they hit the same gene/operon
- **Late**: Mutations are spread more evenly across different targets

**Conclusion**: This supports the module-selection balance prediction — early evolution targets a few high-value genes far from their optimum, while later evolution spreads across many targets with more similar selection coefficients.

# Part 4: Shannon Entropy Analysis

**Shannon entropy** measures the "uncertainty" or "information content" of the distribution.

- **Higher entropy** = mutations spread evenly across many targets
- **Lower entropy** = mutations concentrated in fewer targets

Entropy is less sensitive than Simpson's D to highly-hit genes, but provides a complementary view.

```
def shannon_entropy(labels):
    """Calculate Shannon entropy in bits."""
    counts = pd.Series(labels).value_counts()
    proportions = counts / counts.sum()
    return scipy_entropy(proportions, base=2)

# Calculate entropy
early_gene_H = shannon_entropy(early['gene'])
late_gene_H = shannon_entropy(late['gene'])
early_operon_H = shannon_entropy(early_op['operon'])
late_operon_H = shannon_entropy(late_op['operon'])

print("Shannon Entropy (bits):")
print(f"  Gene-level:   Early = {early_gene_H:.2f}, Late = {late_gene_H:.2f}")
print(f"  Operon-level: Early = {early_operon_H:.2f}, Late = {late_operon_H:.2f}")

# Permutation test for entropy
obs_diff_gene_H = early_gene_H - late_gene_H
obs_diff_operon_H = early_operon_H - late_operon_H

perm_diffs_gene_H = []
perm_diffs_operon_H = []

np.random.seed(42)
for _ in range(n_perm):
    shuffled = np.random.permutation(all_genes)
    perm_diffs_gene_H.append(shannon_entropy(shuffled[:n_early]) - shannon_entropy(shuffled[n_early:]))
    
    shuffled_op = np.random.permutation(all_operons)
    perm_diffs_operon_H.append(shannon_entropy(shuffled_op[:n_early_op]) - shannon_entropy(shuffled_op[n_early_op:]))

perm_diffs_gene_H = np.array(perm_diffs_gene_H)
perm_diffs_operon_H = np.array(perm_diffs_operon_H)

# P-values (early should have LOWER entropy, so diff should be negative)
p_gene_H = np.mean(perm_diffs_gene_H <= obs_diff_gene_H)
p_operon_H = np.mean(perm_diffs_operon_H <= obs_diff_operon_H)

print(f"\nPermutation test (n={n_perm}):")
print(f"  Gene-level:   observed diff = {obs_diff_gene_H:.2f}, p = {p_gene_H:.4f}")
print(f"  Operon-level: observed diff = {obs_diff_operon_H:.2f}, p = {p_operon_H:.4f}")
```

```
# Plot entropy results
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# Top left: Gene entropy
ax1 = axes[0, 0]
bars1 = ax1.bar([f'Early\nn={len(early)}', f'Late\nn={len(late)}'], 
                [early_gene_H, late_gene_H], color=['steelblue', 'darkorange'])
ax1.set_ylabel("Shannon Entropy (bits)")
ax1.set_title('Gene-Level')
for bar, val in zip(bars1, [early_gene_H, late_gene_H]):
    ax1.text(bar.get_x() + bar.get_width()/2, val + 0.05, f'{val:.2f}', ha='center', fontsize=11)

# Top right: Operon entropy
ax2 = axes[0, 1]
bars2 = ax2.bar([f'Early\nn={len(early_op)}', f'Late\nn={len(late_op)}'], 
                [early_operon_H, late_operon_H], color=['steelblue', 'darkorange'])
ax2.set_ylabel("Shannon Entropy (bits)")
ax2.set_title('Operon-Level')
for bar, val in zip(bars2, [early_operon_H, late_operon_H]):
    ax2.text(bar.get_x() + bar.get_width()/2, val + 0.05, f'{val:.2f}', ha='center', fontsize=11)

# Bottom: Permutation nulls
ax3 = axes[1, 0]
ax3.hist(perm_diffs_gene_H, bins=50, color='gray', alpha=0.7, edgecolor='black')
ax3.axvline(x=obs_diff_gene_H, color='red', linewidth=2, label=f'Observed = {obs_diff_gene_H:.2f}')
ax3.set_xlabel("Difference in Entropy (Early - Late)")
ax3.set_ylabel('Count')
ax3.set_title(f'Permutation Test (p = {p_gene_H:.4f})')
ax3.legend()

ax4 = axes[1, 1]
ax4.hist(perm_diffs_operon_H, bins=50, color='gray', alpha=0.7, edgecolor='black')
ax4.axvline(x=obs_diff_operon_H, color='red', linewidth=2, label=f'Observed = {obs_diff_operon_H:.2f}')
ax4.set_xlabel("Difference in Entropy (Early - Late)")
ax4.set_ylabel('Count')
ax4.set_title(f'Permutation Test (p = {p_operon_H:.4f})')
ax4.legend()

plt.suptitle("Shannon Entropy: Early vs Late Mutations\n(Higher entropy = more dispersed)", fontsize=14)
plt.tight_layout()
plt.show()
```

## Why is entropy less significant than Simpson's D?

Shannon entropy is more sensitive to the **number of categories** (genes hit), while Simpson's D is more sensitive to **concentration in top categories** (genes hit many times).

Early evolution has a few genes hit 10+ times (malT, hslU, pykF). Simpson's D captures this "a few genes dominate" pattern because squaring amplifies the contribution of highly-hit genes. Entropy treats each gene more equally.

We tested whether beneficial mutations become more dispersed across genes/operons over evolutionary time, as predicted by the module-selection balance theory.

**Simpson's D** (probability two random mutations hit the same target):
- Gene-level: Early = {early_gene_D:.4f}, Late = {late_gene_D:.4f} (p < 0.0001)
- Operon-level: Early = {early_operon_D:.4f}, Late = {late_operon_D:.4f} (p < 0.0001)

## Interpretation

Early mutations are significantly more concentrated in fewer genes/operons, while late mutations spread more evenly across targets. This is consistent with module-selection balance: early evolution targets a few high-value genes far from their optimum, while later evolution spreads across many targets with more similar selection coefficients.

The permutation tests confirm that this pattern is not simply due to the different sample sizes between early and late periods.
