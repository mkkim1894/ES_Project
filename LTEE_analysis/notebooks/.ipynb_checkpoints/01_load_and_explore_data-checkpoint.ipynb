# LTEE Mutation Temporal Analysis

## Project Goal
Test the module-selection balance prediction: beneficial mutations should be concentrated in few genes early in evolution and become more dispersed later.

## Data Source
- **Paper**: Good BH, McDonald MJ, Barrick JE, Lenski RE, Desai MM. "The Dynamics of Molecular Evolution Over 60,000 Generations." *Nature* 551:45-50 (2017)
- **Repository**: https://github.com/benjaminhgood/LTEE-metagenomic
- **Reference genome**: REL606 (NC_012967.1)

## Environment
- **Python**: 3.10
- **Conda environment**: `ltee`
- **Key packages**: pandas, numpy, matplotlib, seaborn, biopython

## Populations
Six nonmutator populations: Ara−5, Ara−6, Ara+1, Ara+2, Ara+4, Ara+5

## Imports and Setup

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
import sys

# Add the LTEE repository to path (to use their parsing functions if needed)
LTEE_REPO_PATH = Path('../LTEE-metagenomic')
sys.path.insert(0, str(LTEE_REPO_PATH))

# Set up paths
DATA_PATH = Path('../data')
LTEE_DATA_PATH = LTEE_REPO_PATH / 'data_files'
LTEE_ADDITIONAL_PATH = LTEE_REPO_PATH / 'additional_data'

print("Paths configured:")
print(f"  LTEE data: {LTEE_DATA_PATH}")
print(f"  Additional data: {LTEE_ADDITIONAL_PATH}")
```

## Define the Six Nonmutator Populations

```python
# The six nonmutator populations (from parse_file.py)
NONMUTATOR_POPULATIONS = ['m5', 'm6', 'p1', 'p2', 'p4', 'p5']

# Population name mapping
def get_pretty_name(pop):
    """Convert internal name to paper name (e.g., 'm5' -> 'Ara-5')"""
    if pop[0] == 'm':
        return f'Ara-{pop[1]}'
    else:
        return f'Ara+{pop[1]}'

print("Nonmutator populations:")
for pop in NONMUTATOR_POPULATIONS:
    print(f"  {pop} -> {get_pretty_name(pop)}")
```

## Load Gene Convergence Matrix

```python
def load_hmm_states(population):
    """Load HMM state trajectories for a population"""
    filepath = LTEE_DATA_PATH / f'{population}_well_mixed_state_timecourse.txt'
    
    with open(filepath, 'r') as f:
        lines = f.readlines()
    
    # First line is timepoints
    timepoints = [int(float(x)) for x in lines[0].strip().split(', ')]
    
    # Lines 5+ are per-mutation state trajectories
    states = []
    for line in lines[5:]:
        vals = [int(float(x)) for x in line.strip().split(', ') if x]
        states.append(vals)
    
    return timepoints, states

def load_mutations(population):
    """Load annotated timecourse for a population"""
    filepath = LTEE_DATA_PATH / f'{population}_annotated_timecourse.txt'
    df = pd.read_csv(filepath, header=0)
    df.columns = df.columns.str.strip()
    
    # Strip whitespace from the Passed? column and filter
    df['Passed?'] = df['Passed?'].str.strip()
    df = df[df['Passed?'] == 'PASS'].copy()
    
    return df

# Test it
pop = 'm5'
mutations_df = load_mutations(pop)
print(f"Loaded {len(mutations_df)} PASS mutations from {pop}")
print(f"\nFirst few rows:")
print(mutations_df[['Position', 'Gene', 'Allele', 'Annotation']].head(10))
```

## Load Annotated Timecourse Data for Each Population


```python
def load_annotated_timecourse(population):
    """
    Load the annotated timecourse file for a population.
    This contains all mutations with their appearance times.
    
    Returns a list of mutation records.
    """
    filename = LTEE_DATA_PATH / f'annotated_timecourse_{population}.txt'
    
    if not filename.exists():
        print(f"File not found: {filename}")
        return None
    
    mutations = []
    with open(filename, 'r') as f:
        for line in f:
            # Parse the line - format depends on file structure
            # We'll need to examine actual file to determine exact format
            mutations.append(line.strip())
    
    return mutations

# Test loading one population
test_pop = 'm5'
test_data = load_annotated_timecourse(test_pop)
if test_data:
    print(f"Loaded {len(test_data)} lines from {test_pop}")
    print("\nFirst few lines:")
    for line in test_data[:5]:
        print(line[:150])  # Print first 150 chars
```
