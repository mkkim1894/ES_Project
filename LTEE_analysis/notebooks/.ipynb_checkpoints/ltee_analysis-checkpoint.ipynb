# LTEE Mutation Temporal Analysis

**Data**: Good et al. (2017) *Nature* 551:45-50. Six nonmutator LTEE populations.

**Repository**: https://github.com/benjaminhgood/LTEE-metagenomic

## Setup and Data Loading

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path

LTEE_REPO_PATH = Path('../LTEE-metagenomic')
LTEE_DATA_PATH = LTEE_REPO_PATH / 'data_files'

NONMUTATOR_POPULATIONS = ['m5', 'm6', 'p1', 'p2', 'p4', 'p5']
T_STAR = 17500

# Colors for early/late distinction
c_early = '#0077BB'
c_late = '#EE7733'
```

```python
def load_hmm_states(population):
    filepath = LTEE_DATA_PATH / f'{population}_well_mixed_state_timecourse.txt'
    with open(filepath, 'r') as f:
        lines = f.readlines()
    timepoints = [int(float(x)) for x in lines[0].strip().split(', ')]
    states = []
    for line in lines[5:]:
        vals = [int(float(x)) for x in line.strip().rstrip(',').split(', ') if x]
        states.append(vals)
    return timepoints, states

def get_appearance_time(state_trajectory, timepoints):
    for i, state in enumerate(state_trajectory):
        if state > 0:
            return timepoints[i]
    return None

def build_mutation_table(populations):
    all_data = []
    for pop in populations:
        filepath = LTEE_DATA_PATH / f'{pop}_annotated_timecourse.txt'
        full_df = pd.read_csv(filepath, header=0)
        full_df.columns = full_df.columns.str.strip()
        full_df['Passed?'] = full_df['Passed?'].str.strip()
        timepoints, states = load_hmm_states(pop)

        # NOTE: Do NOT reset_index. We keep the original row index so that
        # states[i] correctly corresponds to the i-th mutation in full_df
        # (the HMM state file has one row per mutation in full_df order).
        pass_df = full_df[full_df['Passed?'] == 'PASS']

        for i, row in pass_df.iterrows():
            if i >= len(states):
                continue
            all_data.append({
                'population': pop,
                'position': row['Position'],
                'gene': row['Gene'].strip() if isinstance(row['Gene'], str) else row['Gene'],
                'annotation': row['Annotation'].strip() if isinstance(row['Annotation'], str) else row['Annotation'],
                'appearance_time': get_appearance_time(states[i], timepoints)
            })
    return pd.DataFrame(all_data)

def simpson_d(labels):
    counts = pd.Series(labels).value_counts()
    proportions = counts / counts.sum()
    return (proportions ** 2).sum()

# Load data; exclude intergenic mutations (no unambiguous gene assignment)
master_df = build_mutation_table(NONMUTATOR_POPULATIONS)
df = master_df[master_df['gene'] != 'intergenic'].copy()
print(f"Total genic mutations: {len(df)}")
```

## Early vs Late Split (for genomic position panel)

```python
early = df[df['appearance_time'] <= T_STAR].copy()
late  = df[df['appearance_time'] >  T_STAR].copy()

print(f"Early (<=T_STAR={T_STAR} gen): {len(early)} mutations in {early['gene'].nunique()} genes")
print(f"Late  (> T_STAR={T_STAR} gen): {len(late)} mutations in {late['gene'].nunique()} genes")
```

## Sliding Window Analysis

```python
window_size = 10000
step_size   = 2500
max_time    = int(df['appearance_time'].max())

window_results = []
for t_start in np.arange(0, max_time - window_size + step_size, step_size):
    t_end    = t_start + window_size
    t_center = t_start + window_size / 2

    w = df[(df['appearance_time'] > t_start) & (df['appearance_time'] <= t_end)]

    if len(w) >= 20:
        window_results.append({
            't_center': t_center,
            'n': len(w),
            'D_gene': simpson_d(w['gene'])
        })

window_df = pd.DataFrame(window_results)
print(window_df[['t_center', 'n', 'D_gene']].to_string())
```

## Rarefaction Analysis

```python
n_subsample  = int(window_df['n'].min())
n_bootstrap  = 1000

print(f"Rarefaction: subsampling to n={n_subsample} mutations per window")

rarefaction_results = []
np.random.seed(42)

for t_start in np.arange(0, max_time - window_size + step_size, step_size):
    t_end    = t_start + window_size
    t_center = t_start + window_size / 2

    w = df[(df['appearance_time'] > t_start) & (df['appearance_time'] <= t_end)]

    if len(w) >= n_subsample:
        d_gene_samples = []
        for _ in range(n_bootstrap):
            subsample = w.sample(n=n_subsample, replace=False)
            d_gene_samples.append(simpson_d(subsample['gene']))

        rarefaction_results.append({
            't_center': t_center,
            'n_original': len(w),
            'D_gene_mean': np.mean(d_gene_samples),
            'D_gene_lo':   np.percentile(d_gene_samples, 2.5),
            'D_gene_hi':   np.percentile(d_gene_samples, 97.5),
        })

rarefaction_df = pd.DataFrame(rarefaction_results)
print(f"Windows included in rarefaction: {len(rarefaction_df)}")
```

---
## Figure 6

**Panel A**: Rarefied sliding-window effective number of gene targets (1/D) over time.  
**Panel B**: Genomic position distribution of early vs late mutations.

```python
fig, axes = plt.subplots(
    2, 1,
    figsize=(7, 8),
    gridspec_kw={'height_ratios': [1, 1]}
)

# ── Panel A: rarefied sliding window (gene-level) ────────────────────────────
ax = axes[0]

t_gene       = rarefaction_df['t_center'] / 1000
# Invert D to get effective number of targets (1/D).
# Inversion reverses the CI: the percentile that gives the LARGEST D
# gives the SMALLEST 1/D, i.e., the lower confidence bound on 1/D.
invD_mean = 1.0 / rarefaction_df['D_gene_mean']
invD_lo   = 1.0 / rarefaction_df['D_gene_hi']   # lower CI bound after inversion
invD_hi   = 1.0 / rarefaction_df['D_gene_lo']   # upper CI bound after inversion

ax.fill_between(t_gene, invD_lo, invD_hi, color='gray', alpha=0.3, label='95% CI')
ax.plot(t_gene, invD_mean, 'o-', color='gray', markersize=5, linewidth=1.5)
ax.axvline(x=T_STAR / 1000, color='black', linestyle='--', alpha=0.6)

ax.set_xlabel('Time (thousands of generations)', fontsize=11)
ax.set_ylabel('Effective number of targets (1/D)', fontsize=11)
ax.set_title(f'Gene level (rarefied, n={n_subsample} mutations per window)', fontsize=11)
ax.text(-0.10, 1.05, 'A', transform=ax.transAxes, fontsize=14, fontweight='bold', va='top')

# ── Panel B: genomic position distribution ───────────────────────────────────
ax = axes[1]

bins = np.linspace(0, 4.7e6, 81)
ax.hist(
    early['position'], bins=bins, density=True,
    color=c_early, alpha=0.6, edgecolor='white', linewidth=0.5,
    label=f'Early (n={len(early)})'
)
ax.hist(
    late['position'], bins=bins, density=True,
    color=c_late,  alpha=0.6, edgecolor='white', linewidth=0.5,
    label=f'Late (n={len(late)})'
)
ax.set_xlabel('Genomic position (Mb)', fontsize=11)
ax.set_ylabel('Density', fontsize=11)
ax.set_title('Genomic distribution of mutations', fontsize=11)
ax.legend(frameon=False, fontsize=10, loc='upper right')
ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'{x/1e6:.1f}'))
ax.text(-0.10, 1.05, 'B', transform=ax.transAxes, fontsize=14, fontweight='bold', va='top')

plt.tight_layout()
plt.savefig('figure6_ltee.pdf', dpi=300, bbox_inches='tight')
plt.show()
print("Saved figure6_ltee.pdf")
```
